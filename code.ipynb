```python
df_country=pd.read_excel('country-code.xlsx')
df_country.head()
```

```python
# to combine the two dataframes
df_final=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
country_names=df_final.Country.value_counts()
```

```python
country_names=df_final.Country.value_counts().index
```

```python
country_val=df_final.Country.value_counts().values
```

```python
# pie chart for top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observations:
1. The top 3 countries in the dataset are USA, China, and India.
2. The pie chart shows that USA accounts for the largest share of observations (approximately 20.5%), followed by China (18.7%) and India (17.2%).
3. The remaining countries collectively account for about 43.6% of observations.
4. This distribution suggests a concentration of data points in a few key countries, which may be useful for further analysis or modeling decisions.

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

observations:
1. The dataset contains 9551 rows and 21 columns.
2. The dataset contains information about restaurants, including their location, cuisine, ratings, and other details.
3. There are some missing values in the dataset, particularly in the 'Cuisines' and 'Rating color' columns.
4. The top three countries with the most restaurants in the dataset are India, United States, and United Kingdom.
5. The majority of the restaurants have an aggregate rating of 4.0, followed by 3.5 and 4.5.
6. The most common rating color is 'Green', which corresponds to a rating of 4.0 and above, indicating that most restaurants in the dataset are well-rated.
7. The 'Cuisines' column has a wide variety of cuisines, with some restaurants offering multiple cuisines.
8. The 'Average Cost for two' column shows a wide range of costs, indicating that the dataset includes restaurants of varying price levels.
9. when rating is between 4.0 and 4.9, the rating color is green and the rating text is 'Excellent'.
